```python
import posixpath
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
data_root = "D:/OneDrive/Documents/Cours/4A/SFE/data/"

kh_fronts = gpd.read_file(posixpath.join(data_root, "KH-5_ARGON_fronts/KH-5_ARGON_fronts.shp")).to_crs(epsg=25833)
landsat_fronts = gpd.read_file(posixpath.join(data_root, "Landsat_MSS_fronts/Landsat_MSS_fronts.shp")).to_crs(epsg=25833)
all_fronts = pd.concat([kh_fronts, landsat_fronts], ignore_index=True)
all_fronts.head()
```

```python
link = pd.DataFrame(
    [],
    columns = ["front_id", "front_dir", "RGI60ID", "RGI70ID"]
)

link.loc[:, "front_id"] = all_fronts.front_id.unique()
link
```

```python
link.loc[:, "front_dir"] = link.front_id.apply(lambda i: all_fronts.loc[all_fronts.front_id == i, "front_dir"].values[0])
link.loc[:, "RGI60ID"] = link.front_id.apply(lambda i: all_fronts.loc[all_fronts.front_id == i, "RGI60ID"].values[0])
link.loc[:, "RGI70ID"] = link.front_id.apply(lambda i: all_fronts.loc[all_fronts.front_id == i, "RGI70ID"].values[0])
link.head()
```

```python
data_root = "D:/OneDrive/Documents/Cours/4A/SFE/data/"

boxes = gpd.read_file(posixpath.join(data_root, "boxes/boxes.shp"), columns=["front_id"]).to_crs(epsg=25833)
boxes.columns = ["front_id_float", 'geometry']
boxes.loc[:, "front_id"] = boxes.front_id_float.astype(int)
boxes = boxes.loc[:, ["front_id", "geometry"]]
boxes
```

```python
boxes.loc[:, "front_dir"] = boxes.front_id.apply(lambda i: link.loc[link.front_id == i].front_dir.values[0])
boxes.loc[:, "RGI60ID"] = boxes.front_id.apply(lambda i: link.loc[link.front_id == i].RGI60ID.values[0])
boxes.loc[:, "RGI70ID"] = boxes.front_id.apply(lambda i: link.loc[link.front_id == i].RGI70ID.values[0])
boxes.loc[:, ["front_id", "RGI60ID", "RGI70ID", "front_dir", "geometry"]]
```
